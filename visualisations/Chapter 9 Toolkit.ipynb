```python
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mplines
import matplotlib.patches as patches



#Animations
from matplotlib import animation, rc
from IPython.display import HTML, Image
```

```python
r = 3
theta = np.linspace(0,2*np.pi,200)
arrowlength = 1.3
phi = theta+np.pi/2

sample_point = 35
sample_pointb =sample_point/2

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.figure(figsize=(5,5))

ax = plt.axes()
ax.set_ylabel('')
ax.set_xlabel('')


ax.axvline(x=0,linewidth=0.5,color='k',linestyle="-")
ax.axhline(y=0,linewidth=0.5,color='k',linestyle="-")

#plt.axvline(x=0,linewidth=0.5,color='r',linestyle="--")
#plt.axhline(y=0,linewidth=0.5,color='b',linestyle="--")
#plt.text(10, 4,'Phase shift, $\delta = $'+str(delta_2), ha="right", va="center", size=10)
#ax.plot(x, y,color='blue')
ax.set_ylim(-0.5,3.5)
ax.set_xlim(-0.5,3.5)
#plt.plot(t, shm(t,delta_2),color='orange')
#ax.set_size(5,5)

# ax.annotate('v', xy=(x[sample_point],y[sample_point]), xytext=((x[sample_point]+arrowlength*np.cos(phi[sample_point])),y[sample_point]+arrowlength*np.sin(phi[sample_point])), arrowprops=dict(arrowstyle='<-'),size = 20)
ax.annotate('', xy=(0,0), xytext=((x[sample_point]),y[sample_point]), arrowprops=dict(arrowstyle='<-'))
ax.text(0.5*x[sample_point]-0.2, 0.5*y[sample_point]+0.2, "r",  rotation=0,size=20 )
ax.plot(x[sample_point],y[sample_point],'ro',markersize='10')
ax.text(1.1*x[sample_point],1.1*y[sample_point],"P",size='20')

ax.text(0.1*x[int(sample_pointb)], 0.1*y[int(sample_pointb)], "$θ$",  rotation=0,size=20)

ax.text(-0.4, 3, "Im($y$)",  rotation=90,size=20)
ax.text(3, -0.4, "Re($x$)",  rotation=0,size=20)


ax.plot([x[sample_point], x[sample_point]], [0, y[sample_point]], color='b', linestyle='--', linewidth=1)
ax.plot([0, x[sample_point]], [y[sample_point], y[sample_point]], color='orange', linestyle='--', linewidth=1)


ax.text(0.5*x[sample_point],-0.2,  "$x$",  rotation=0,size=15)
ax.text(x[sample_point]+0.2, 0.5*y[sample_point], "$y$",  rotation=0,size=15)


ax.set_xticklabels([])
ax.set_yticklabels([])


ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')


plt.tight_layout()
plt.savefig('ch9-complexnumbers1.pdf',transparent=True)
plt.savefig('ch9-complexnumbers1.png',transparent=True)
plt.savefig('ch9-complexnumbers1.svg',transparent=True)
plt.show()
```

```python
r = 3
theta = np.linspace(0,2*np.pi,200)
arrowlength = 1.3
phi = theta+np.pi/2

sample_pointa = 35
sample_pointb = 15

ascale = 0.5
bscale = 1.2

x = r*np.cos(theta)
y = r*np.sin(theta)

z1 = [ascale*x[sample_pointa],ascale*y[sample_pointa]]
z2 = [bscale*x[sample_pointb],bscale*y[sample_pointb]]



plt.figure(figsize=(6,5))

ax = plt.axes()
ax.set_ylabel('')
ax.set_xlabel('')

ax.set_ylim(-1,(x[sample_pointa]+x[sample_pointb]))
ax.set_xlim(-1,(y[sample_pointa]+y[sample_pointb]+1))

# ax.annotate('v', xy=(x[sample_point],y[sample_point]), xytext=((x[sample_point]+arrowlength*np.cos(phi[sample_point])),y[sample_point]+arrowlength*np.sin(phi[sample_point])), arrowprops=dict(arrowstyle='<-'),size = 20)
#ax.text(0.5*x[sample_point]-0.2, 0.5*y[sample_point]+0.2, "r",  rotation=0,size=20 )

ax.text(-0.4, 3, "Im($y$)",  rotation=90,size=20)
ax.text(3, -0.4, "Re($x$)",  rotation=0,size=20)


# z1
ax.plot(z1[0],z1[1],'ro',markersize='10')
#ax.annotate('', xy=(0,0), xytext=((x[sample_pointa]),y[sample_pointa]), arrowprops=dict(arrowstyle='<-'))
ax.text(1.1*z1[0],1.1*z1[1],"$z_1$",size='20')
ax.plot([0,z1[0]], [0, z1[1]], color='b', linestyle='-', linewidth=1)



# z2:
ax.plot(z2[0],z2[1],'ro',markersize='10')
#ax.annotate('', xy=(0,0), xytext=((x[sample_pointb]),y[sample_pointb]), arrowprops=dict(arrowstyle='<-'))
ax.text(1.1*z2[0],1.1*z2[1],"$z_2$",size='20')
ax.plot([0,z2[0]], [0, z2[1]], color='b', linestyle='-', linewidth=1)



# z1+z2
ax.plot(z1[0]+z2[0],z1[1]+z2[1],'ko',markersize='10')
ax.plot([0,z1[0]+z2[0]], [0, z1[1]+z2[1]], color='k', linestyle='-', linewidth=1)
ax.text(1.05*(z1[0]+z2[0]),1.05*(z1[1]+z2[1]),"$z_1 + z_2$",size='20')



# z2 -> z1+z2
ax.plot([z2[0],z1[0]+z2[0]], [z2[1], z1[1]+z2[1]], color='orange', linestyle='--', linewidth=1)
ax.plot([z1[0],z1[0]+z2[0]], [z1[1], z1[1]+z2[1]], color='orange', linestyle='--', linewidth=1)






#ax.plot([x[sample_point], x[sample_point]], [0, y[sample_point]], color='b', linestyle='--', linewidth=1)
#ax.plot([0, x[sample_point]], [y[sample_point], y[sample_point]], color='orange', linestyle='--', linewidth=1)


ax.set_xticklabels([])
ax.set_yticklabels([])


ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')

plt.tight_layout()
plt.savefig('ch9-complexnumbers2.pdf',transparent=True)
plt.savefig('ch9-complexnumbers2.png',transparent=True)
plt.savefig('ch9-complexnumbers2.svg',transparent=True)

plt.show()
```
