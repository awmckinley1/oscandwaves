```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML, Image

rc('animation', html='html5') # Needed for in-line movie
```

```python
# via https://stackoverflow.com/questions/9215658/plot-a-circle-with-pyplot

list_of_circs=np.arange(0.1,2.5,0.35)
circle1 = plt.Circle((0, 0), 0.04, color='k',fill=True)
circle2 = plt.Circle((-0.7, 0), 0.04, color='k',fill=True)


size_of_font=15
integer_list = np.arange(1,len(list_of_circs)+1,1)
integer_flip = np.flip(integer_list,axis=0)
integers = (["7","6","5","4","3","2","2"])

fig, ax = plt.subplots(figsize=(7,7)) # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()

ax.set(xlim=(-2.8,1.2),ylim=(-2,2))
ax.set_aspect('equal')

for i in list_of_circs:
    circle=plt.Circle((0-0.6*i, 0), i, color='b',fill=False, alpha = 0.8*(1-(2.5*i/len(list_of_circs)))**2,linewidth=3)
    centre=plt.Circle((0-0.6*i, 0), 0.03, color='r',fill=True, alpha = 0.8*(1-(2.5*i/len(list_of_circs)))**2)
    ax.add_artist(circle)
    ax.add_artist(centre)
    index_i = np.where(list_of_circs == i)
    ax.annotate(int(integer_flip[index_i]),xy=(0-0.6*i, -0.2),color='k',fontsize=size_of_font,ha='center')
    ax.annotate(int(integer_flip[index_i]),xy=(0-0.6*i, +i+0.05),color='k',fontsize=size_of_font,ha='center')

ax.annotate("", xy=(1, 0), xytext=(0.1, 0),arrowprops=dict(facecolor='black'))
ax.annotate("$u_s$",xy = (1, 0),fontsize=20)

ax.annotate("", xy=(0.5, -1.2), xytext=(0.151, -0.851),arrowprops=dict(facecolor='blue',))
ax.annotate("", xy=(0.5, 1.2), xytext=(0.151, 0.851),arrowprops=dict(facecolor='blue',))


ax.annotate("$v$",xy = (0.55, -1.2),fontsize=20)
ax.annotate("$v$",xy = (0.55, 1.2),fontsize=20)





    
#for i in list_of_circs:
#    circle=plt.Circle((0, 0), i, color='r',fill=False, alpha = (1-(5*i/len(list_of_circs))))
#    ax.add_artist(circle)

ax.add_artist(circle1)
ax.axis('off')
plt.tight_layout()


plt.savefig('ch9-doppler2.pdf',transparent=True)
plt.savefig('ch9-doppler2.svg',transparent=True)




plt.show()
```
